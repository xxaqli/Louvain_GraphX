```python
import toyplot
import numpy as np
inputPath = '/Users/binli/Desktop/unionpay/GraphX/Project/louvain_community_detection'
edgesFile = open(inputPath+'/data/club.txt','r')
edges = []
for lines in edgesFile:
    (src,dst) = lines.strip('\r\n').split(' ')
    edges.append((int(src),int(dst)))
edges_array = np.array(edges)
toyplot.graph(edges_array)
```

```python
communityFile = open(inputPath+'/data/community/part-00000','r')
count = 0
truth_community = []
for lines in communityFile:
    count += 1
    cid,vertice_str = lines.strip('\r\n').split('|-')
    vertice_list = vertice_str.split(' ')
    for vertice in vertice_list:
        vertice_code = (int(vertice),int(count))
        truth_community.append(vertice_code)
truth_community = sorted(truth_community,key=lambda x:x[0])  #第一列数据需要从小到大排序
community_array = np.array(truth_community)
colormap = toyplot.color.brewer.map("Set2")
vcolor = community_array[:,1]
ecolor = "lightgray"
vlstyle = {"fill":"white"}
canvas, axes, mark = toyplot.graph(edges_array, ecolor=ecolor, vsize=20, vlstyle=vlstyle, vcolor=(vcolor, colormap))
```

```python
import toyplot.pdf
toyplot.pdf.render(canvas, "louvain_communities.pdf")
```
